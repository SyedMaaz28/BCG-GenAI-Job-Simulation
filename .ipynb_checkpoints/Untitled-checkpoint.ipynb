### Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Reading the Manually Extracted Dataset from SEC's EDGAR's database


### https://www.sec.gov/edgar/browse/?CIK

```python
data = pd.read_csv("Fiscal_Years _Data_Report.csv")
```

```python
data
```

### Calculating Year-by-Year growth rates for Total Revenue and Net Income


```python
data['Revenue Growth (%)'] = data.groupby(['Company'])['Total Revenue'].pct_change() * 100
data['Net Income Growth (%)'] = data.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
data
```

### Imputing Null / NaN values that result from pct_change to 0

```python
data.fillna(0,inplace = True)
```

```python
data
```

### Calculating Year-by-Year Growth for Total Assets , Total Libalities and Cash Flow

```python
data['Assets Growth (%)'] = data.groupby('Company')['Total Assets'].pct_change() * 100
data['Liabilities Growth (%)'] = data.groupby('Company')['Total Liabilities'].pct_change() * 100
data['Cash Flow from Operations Growth(%)'] = data.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
data
```

### Imputing Null / NaN values with 0

```python
data.fillna(0 , inplace=True)
```

```python
data
```

### Creating Plots 

```python
plt.figure(figsize=(12, 6))
import warnings
warnings.filterwarnings("ignore")
```

```python
# Creating a bar plot for Total Revenue
sns.barplot(
    data=data, 
    x="Year", 
    y="Total Revenue", 
    hue="Company", 
    ci=None, 
    palette="viridis"
)
plt.title("Total Revenue by Company and Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Revenue (in millions)", fontsize=12)
plt.legend(title="Company")
plt.grid(axis="y", linestyle="--", alpha=0.3)
plt.show()
```

```python
# Plotting Net Income Growth for each company by year

sns.lineplot(
    data=data, 
    x="Year", 
    y="Net Income Growth (%)", 
    hue="Company", 
    marker="o", 
    palette="deep"
)
plt.title("Net Income Growth by Company and Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Net Income Growth (%)", fontsize=12)
plt.legend(title="Company")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
